```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df_numeric = pd.read_csv("df_numeric.csv")
df_area = df_numeric[['price','area']]
df_area
```

```python
X = df_area[['area']]
Y = df_area['price']
```

```python
lm = LinearRegression() # crea el modelo y luego ponemos sus parametros
lm.fit(X,Y)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
lm.score(X,Y)
```

#### That R2 was pretty good for a first try!


```python
#Moving on to a multiple linear regression, we will be including 'bedrooms','bathrooms', and garage.

#To do that, we need to create dummies, since these are categorical. 
bd = pd.get_dummies(df_numeric['bedrooms'],"bd",drop_first=True)
bt = pd.get_dummies(df_numeric['bathrooms'],"bt",drop_first=True)
grg= pd.get_dummies(df_numeric['garage'],"grg",drop_first=True)

#then we merge the dummies into a single df called df_dummies. 
df_dummies = pd.concat([df_numeric,bd],axis=1)
df_dummies = pd.concat([df_dummies,bt],axis=1)
df_dummies = pd.concat([df_dummies,grg],axis=1)
df_dummies
```

```python
X = df_dummies[['area','bd_1', 'bd_2', 'bd_3', 'bd_4', 'bd_5', 'bd_6', 'bt_2', 'bt_3', 'bt_4','grg_1','grg_2','grg_3','grg_4']]
Y = df_dummies['price']
```

```python
lm = LinearRegression() # crea el modelo y luego ponemos sus parametros
lm.fit(X,Y)
```

```python
print(lm.intercept_)
print(lm.coef_)
```

```python
lm.score(X,Y)
```
