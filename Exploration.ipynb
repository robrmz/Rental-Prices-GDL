```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt


#Set how many float values
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

```python
#Open, set index and print
df = pd.read_csv('rental_data_formatted.csv')
#Dropping missing values
#df = df.dropna()
df = df.set_index('Unnamed: 0')
```

```python
#Setting a sensible float number

df['price'].describe()
plt.boxplot(df['price'],meanline=True,notch=False)
plt.show()
```

```python
print(df['bedrooms'].describe())
print('\n')
print(df['bathrooms'].describe())
print('\n')
print(df['garage'].describe())
print('\n')

print(df.boxplot(column=['bedrooms','bathrooms','garage']))
```

```python
df['location'] = df['location'].astype('category').cat.codes
df['location'].value_counts()
```

```python
corr_matrix = df.corr()
sn.heatmap(corr_matrix, annot=True)
plt.show()
```

```python
df.hist(column='bedrooms', grid=False, legend=True)
df.hist(column='price', grid=False, legend=True)
df.hist(column='bathrooms', grid=False, legend=True)
df.hist(column='garage', grid=False, legend=True)
df.hist(column='area', grid=False, legend=True)
```
